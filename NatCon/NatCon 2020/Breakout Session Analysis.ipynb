```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_BS = pd.read_csv('SVA NatCon 2020 Post-Conference Survey.csv',encoding='latin-1')

df_BS.head()
```

```python
def get_breakout_rates(df_new):
    df = df_new
    df.columns = ['title','topic','Length','Usefulness','Presentation','Discussion, Q&A','Overall','Comments']

    title_list = list(set(df['title'].tolist()))

    rates_dict = {}

    for i in range(len(title_list)):
        title = title_list[i]

        df_title = df[df['title'] == title].reset_index()

        value_dict = {'Poor':1,'Fair':2,'Average':3,'Good':4,'Excellent':5}
        topic_lst = []
        len_lst = []
        use_lst = []
        pre_lst =[]
        QA_lst = []
        all_lst = []
        com_lst = []

        for i in range(len(df_title)):
            try:
                topic_lst.append(value_dict[df_title['topic'][i]])
            except:
                pass

            try:
                len_lst.append(value_dict[df_title['Length'][i]])
            except:
                pass

            try:
                use_lst.append(value_dict[df_title['Usefulness'][i]])
            except:
                pass    
            try:
                pre_lst.append(value_dict[df_title['Presentation'][i]])
            except:
                pass
            try:
                QA_lst.append(value_dict[df_title['Discussion, Q&A'][i]])
            except:
                pass
            try:
                all_lst.append(value_dict[df_title['Overall'][i]])
            except:
                pass

            try:
                if len(df_title['Comments'][i])> 1:
                    com_lst.append(df_title['Comments'][i]+'     /next/')
            except:
                pass

        rate_dict = {'Attendance':len(df_title),'topic':np.mean(topic_lst),'Length':np.mean(len_lst),'Usefulness':np.mean(use_lst),'Presentation':np.mean(pre_lst),'Discussion, Q&A':np.mean(QA_lst),'Overall':np.mean(all_lst),'Comments':com_lst}
        rates_dict[title] = rate_dict
    return rates_dict
    
```

```python
df = df_BS[['Q14','Q15_1','Q15_2','Q15_3','Q15_4','Q15_5','Q15_6','Q76']]

rates_dict = get_breakout_rates(df)
rates_dict


I should probably go to bed
chasin you
Lovin on you
circles
```
